```python
import pandas as pd
from plotnine import *
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
# import graphviz 

# importing statsmodels (for linear regression in python)
import statsmodels.formula.api as smf
```

## Load Data

Code for loading raw data like `pd.read_csv()` etc. -- will go into this section. 

```python
df2 = pd.read_excel("ACC HS Chronic Absenteeism.xlsx",dtype={'ENTITY_CD':str})
df3 = pd.read_excel("ACC HS Graduation Rate.xlsx",dtype={'ENTITY_CD':str})
```

```python
df2['charter'] = (df2.ENTITY_CD.str.slice(start=6, stop=8) == '86').astype(int)
df2['nyc'] = df2.ENTITY_CD.str[0] == '3'

# 86 == Charter
# 01 == "CITY SD"???
# 02 == Union Free School Districts
# 03 ==
# 04 ==
# 05 ==
# 06 == CSD???
# 07 ==
# 11 ==
```

### Clean data

```python
# Filtering for the right data points

# absenteeism = df2[df2.SUBGROUP_NAME == 'All Students']
absenteeism = df2
# all_students_df3 = df3[df3.SUBGROUP_NAME != 'All Students']
all_students_df3 = df3
graduation_cohort = all_students_df3[
    (all_students_df3.YEAR == 2021) &
    (all_students_df3.COHORT != 'Combined') &
    (all_students_df3.GRAD_COUNT != 's') &
    (all_students_df3.GRAD_RATE != 's')
]

# Left join to see missing schools
all_data = graduation_cohort.merge(
    absenteeism, 
    on=['INSTITUTION_ID', 'ENTITY_CD', 'ENTITY_NAME', 'YEAR', 'SUBGROUP_NAME'], 
    how='left'
)
```

```python
all_data.dtypes
```

## Getting rid of entries with invalid input ('s')

```python
# Filter for rows where ENROLLMENT, ABSENT_COUNT, and ABSENT_RATE are all not s
df_filtered = all_data[
    (all_data['ENROLLMENT'] != 's') &
    (all_data['ABSENT_COUNT'] != 's') &
    (all_data['ABSENT_RATE'] != 's')
].copy()
```

```python
# Convert INSTITUTION_ID and ENTITY_CD to strings, but convert other columns to numeric
df_filtered.INSTITUTION_ID = df_filtered.INSTITUTION_ID.astype(str)
df_filtered.ENTITY_CD = df_filtered.ENTITY_CD.astype(str)
df_filtered.ENROLLMENT = df_filtered.ENROLLMENT.astype(float)
df_filtered.ABSENT_COUNT = df_filtered.ABSENT_COUNT.astype(float)
df_filtered.ABSENT_RATE = df_filtered.ABSENT_RATE.astype(float)
df_filtered.COHORT_COUNT = df_filtered.COHORT_COUNT.astype(float)
df_filtered.GRAD_COUNT = df_filtered.GRAD_COUNT.astype(float)
df_filtered.GRAD_RATE = df_filtered.GRAD_RATE.astype(float)
```

```python
df_filtered.dtypes
```

```python
df_filtered.SUBGROUP_NAME.unique()
```

```python
#Limiting data to just charter schools
# df_charter = df_filtered[df_filtered.ENTITY_NAME.str.contains("CHARTER")]
```

```python
df_all = df_filtered.query("SUBGROUP_NAME=='All Students'")
df_disadvantaged = df_filtered.query("SUBGROUP_NAME == 'Economically Disadvantaged'")
df_black = df_filtered.query("SUBGROUP_NAME == 'Black or African American'")
df_white = df_filtered.query("SUBGROUP_NAME == 'White'")
df_hisp = df_filtered.query("SUBGROUP_NAME == 'Hispanic or Latino'")
df_english_learners = df_filtered.query("SUBGROUP_NAME == 'English Language Learners'")

# KEEP ADDING THINGS YOU NEED FOR THE REGRESSIONS HERE
dataframes_to_merge = [df_disadvantaged, df_black, df_white, df_hisp, df_english_learners]

df_new = df_all.copy()
for this_df in dataframes_to_merge:
    df_new = df_new.merge(
        this_df[['INSTITUTION_ID', 'ENROLLMENT', 'GRAD_RATE', 'ABSENT_RATE', 'COHORT', 'COHORT_COUNT']], 
        on=['INSTITUTION_ID', 'COHORT'], 
        how='left',
        suffixes =('', '_' + this_df.SUBGROUP_NAME.unique()[0].lower().split(' ')[0]))

df_new['ENROLLMENT_economically'] = df_new['ENROLLMENT_economically'] / df_new['ENROLLMENT'] * 100
df_new['ENROLLMENT_black'] = df_new['ENROLLMENT_black'] / df_new['ENROLLMENT'] * 100
df_new['ENROLLMENT_white'] = df_new['ENROLLMENT_white'] / df_new['ENROLLMENT'] * 100
df_new['ENROLLMENT_hispanic'] = df_new['ENROLLMENT_hispanic'] / df_new['ENROLLMENT'] * 100
df_new['ENROLLMENT_english'] = df_new['ENROLLMENT_english'] / df_new['ENROLLMENT'] * 100

pd.set_option('display.max_columns', None)

df_new.head(50)
```

```python
# EXPLAINING BLACK STUDENTS GRADUATING RATES WITH ECONOMICALLY DISADVANTAGED, ABSENTHEISM RATE AND BEING A CHARTER SCHOOL OR NOT


model = smf.ols("GRAD_RATE_black ~ ENROLLMENT_economically + ABSENT_RATE + charter", data=df_new)
results = model.fit()
results.summary()
```

```python
model = smf.ols("GRAD_RATE_black ~ charter + ENROLLMENT_economically + ", data=df_new)
results = model.fit()
results.summary()
```
